```python
# Data Cleaning dan Preparation
# Exercise 1: Identifying and Handling Missing Data
import pandas as pd

# Sample dataset with missing values
data = {
'Name': ['Alice', 'Bob', 'Charlie', 'David', None],
'Age': [24, 30, None, 22, 35],
'Salary': [48000, None, 57000, None, 60000]
}
df = pd.DataFrame(data)

# Filling missing values and dropping rows
df['Age'].fillna(df['Age'].mean(), inplace=True)
df['Salary'].fillna(df['Salary'].median(), inplace=True)
df.dropna(subset=['Name'], inplace=True)
print('After cleaning:\n', df)
```

```python
# Exercise 2: Standardizing Categorical Data
# Sample dataset with inconsistent categorical values
data = {
    'Product': ['Laptop', 'Laptop', 'Desktop', 'Tablet', 'Tablet'],
    'Category': ['Electronics', 'electronics', 'Electronics', 'Gadgets', 'gadgets']
}
df = pd.DataFrame(data)

# Standardize category values
df['Category'] = df['Category'].str.capitalize()
print('Standardized Data:\n', df)
```

```python
# Tugas Praktik
#  Load Dataset and Identify missing values
import pandas as pd
import numpy as np
from sklearn.preprocessing import MinMaxScaler, StandardScaler

# 1. Load dataset
df = pd.read_csv ('/home/fdkbigdata/Downloads/train.csv')

# 2. Identifikasi missing values
print("Jumlah missing tiap kolom:")
print(df.isna().sum())
```

```python
# Implement data transformation to normalize numerical columns
# 3. Normalisasi kolom numerik
numeric_cols = df.select_dtypes(include=['int64','float64']).columns.tolist()
print("Kolom numerik:", numeric_cols)

# Misalnya kita skip kolom target (“Survived”) jika ada
if 'Survived' in numeric_cols:
    numeric_cols.remove('Survived')

scaler = MinMaxScaler()  # atau StandardScaler()
df[numeric_cols] = scaler.fit_transform(df[numeric_cols])
print("Setelah normalisasi kolom numerik:")
print(df[numeric_cols].head())
```

```python
# Standardize categorical columns and remove duplicates
# 4. Standarisasi kolom kategorikal
categorical_cols = df.select_dtypes(include=['object']).columns.tolist()
print("Kolom kategorikal:", categorical_cols)

# Contoh sederhana: ubah semua ke huruf kecil + kapitalisasi
for col in categorical_cols:
    df[col] = df[col].str.strip().str.lower().str.capitalize()

print("Setelah standarisasi kategori:")
print(df[categorical_cols].head())
```

```python
# 5. Menghapus duplikat
before = df.shape[0]
df = df.drop_duplicates()
after = df.shape[0]
print(f"Baris sebelum drop duplicates: {before}, setelah: {after}")
print(df.head())
```

```python
# Tugas Tambahan
# Clean a real-world dataset, perform normalization, handle outlier, and prepare the data for analysis
import pandas as pd

# Load dataset
df = pd.read_csv('/home/fdkbigdata/Downloads/netflix_titles.csv')

# Lihat 5 baris pertama
df.head()
```

```python
# Cek Informasi Umum dan Missing Values
df.info()
df.isnull().sum()
```

```python
# Data Cleaning
# Hapus duplikasi
df.drop_duplicates(inplace=True)
# Isi missing value pada kolom 'rating' dengan 'Unknown'
df['rating'] = df['rating'].fillna('Unknown')
# Isi missing country dengan 'Unknown'
df['country'] = df['country'].fillna('Unknown')
# Untuk kolom teks seperti 'director' atau 'cast', ganti NaN jadi 'Not Available'
df['director'] = df['director'].fillna('Not Available')
df['cast'] = df['cast'].fillna('Not Available')


# Feature Extraction
# Ubah release_year menjadi integer
df['release_year'] = df['release_year'].astype(int)

# Ubah date_added menjadi format tanggal
df['date_added'] = pd.to_datetime(df['date_added'], errors='coerce')

# Ambil tahun penambahan ke Netflix
df['year_added'] = df['date_added'].dt.year
```

```python
# Handle Outliers (Contoh Kolom Numerik)
import numpy as np
Q1 = df['release_year'].quantile(0.25)
Q3 = df['release_year'].quantile(0.75)
IQR = Q3 - Q1

# Tentukan batas
lower_bound = Q1 - 1.5 * IQR
upper_bound = Q3 + 1.5 * IQR

# Filter outliers (film terlalu lama atau terlalu baru)
df = df[(df['release_year'] >= lower_bound) & (df['release_year'] <= upper_bound)]

```

```python
# Normalization (pada data numerik)
from sklearn.preprocessing import MinMaxScaler

# Buat kolom 'years_since_release'
df['years_since_release'] = 2025 - df['release_year']

# Normalisasi ke rentang 0–1
scaler = MinMaxScaler()
df['years_since_release_norm'] = scaler.fit_transform(df[['years_since_release']])
```

```python
# Data Preparation for Analysis
clean_df = df[['title', 'type', 'country', 'release_year', 'rating',
               'duration', 'listed_in', 'years_since_release_norm']]

clean_df.head()
```
