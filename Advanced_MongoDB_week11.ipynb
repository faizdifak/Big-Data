```python
from pymongo import MongoClient, InsertOne, UpdateOne, DeleteOne

# Koneksi ke MongoDB
client = MongoClient('mongodb://localhost:27017/')
db = client['university_db']
courses_collection = db['courses']

# Hapus semua data sebelumnya biar tidak dobel saat testing
courses_collection.delete_many({})

# --- Latihan 1 dan Tugas 1 ---
operations = [
    # Insert beberapa dokumen course (masing-masing punya daftar siswa)
    InsertOne({
        'course': 'Math 101', 'enrollments': 30, 'department': 'Mathematics','students': [
            {'student_id': 1, 'name': 'Asep'},
            {'student_id': 2, 'name': 'Boby'}
        ]
    }),
    InsertOne({
        'course': 'CS 102', 'enrollments': 25, 'department': 'Computer Science', 'students': [
            {'student_id': 3, 'name': 'Chunha'},
            {'student_id': 4, 'name': 'David'},
            {'student_id': 5, 'name': 'Eva'}
        ]
    }),
    InsertOne({
        'course': 'History 201', 'enrollments': 20, 'department': 'History', 'students': [
            {'student_id': 6, 'name': 'Fiona'}
        ]
    }),
    InsertOne({
        'course': 'Physics 202', 'enrollments': 15, 'department': 'Computer Physics', 'students': [
            {'student_id': 7, 'name': 'Meli'},
            {'student_id': 8, 'name': 'Dani'},
            {'student_id': 9, 'name': 'Heru'}
        ]
    }),

    # Update â€” misal menambah siswa baru di Math 101
    UpdateOne(
        {'course': 'Math 101'},
        {'$push': {'students': {'student_id': 10, 'name': 'Rusdi'}}}
    ),

    # Delete â€” hapus course yang tidak aktif lagi
    DeleteOne({'course': 'Physics 202'})
]

# Jalankan operasi bulk
result = courses_collection.bulk_write(operations)

print("Bulk operasi selesai.")
print(f"Dokumen disisipkan: {result.inserted_count}")
print(f"Dokumen diperbarui: {result.modified_count}")
print(f"Dokumen dihapus: {result.deleted_count}")

# Tampilkan data hasil akhir
print("\nData semua course saat ini:")
for course in courses_collection.find():
    print(course)
```

```python
# Memfilter dan Kueri Kompleks
# Latihan 2
# Filter kursus dengan pendaftaran lebih dari 20 siswa
for course in courses_collection.find({'enrollments':{'$gt': 20}}):
    print(course)
```

```python
# Tugas 2
# Mengambil dan Menampilkan kursus hanya dari jurusan 'ilmu Komputer' atau "Matematika'
for course in courses_collection.find({'department':{'$in': ['Computer Science', 'Matematics']}}):
    print(course)
```

```python
# Kerangka Agregasi untuk Analisi Data
# Latihan 3
# Gunakan Agregasi untuk menemukan rata2 pendaftaran per department
pipeline = [
    {'$group': {'_id': '$department', 'average_enrollment': {'$avg': '$enrollments'}}}
]
for result in courses_collection.aggregate(pipeline):
    print(result)
```

```python
# Tugas 3
# Membuat jalur agregasi yang menemukan pendaftaran maksimum untuk setiap department
pipeline = [
    {'$group': {'_id': '$department', 'max_enrollment': {'$max': '$enrollments'}}}
]
for result in courses_collection.aggregate(pipeline):
    print(result)
```

```python
# Transformasi Data menggunakan Project dan Menambahkan Bidang
#Latihan 4
# Gunakan $project untuk mengganti nama dan hanya menampilkan bidang: course_name, department_name, dan enrollments
pipeline = [
    { '$project' : { 'course_name' : '$course' , 'department_name' : '$department' , 'enrollments' : 1 }}
]
for result in courses_collection.aggregate(pipeline):
    print(result)
```

```python
# Tugas 4
# Gunakan $addFields untuk membuat bidang baru enrollment_category dimana enrollment > 20 high dan yang lainya low
pipeline = [
    {'$addFields' : {'enrollment_category' : {'$cond':{'if': {'$gt': ['$enrollments', 20]}, 'then': 'high', 'else': 'low'}}}}
]
for result in courses_collection.aggregate(pipeline):
    print(result)
```

```python
# Tugas Tambahan
print("\n=== TUGAS 1: Jumlah mata kuliah per departemen ===")
pipeline1 = [
    {
        '$group': {
            '_id': '$department',
            'jumlah_mata_kuliah': {'$sum': 1}
        }
    }
]

for doc in courses_collection.aggregate(pipeline1):
    print(doc)
```

```python
print("\n=== TUGAS 2: Kursus dengan enrollments > 25 di Computer Science ===")
pipeline2 = [
    {
        '$match': {
            'department': 'Computer Science',
            'enrollments': {'$gte': 25}  # gunakan $gte
        }
    },
    {
        '$group': {
            '_id': '$department',
            'total_kursus': {'$sum': 1},
            'daftar_kursus': {'$push': '$course'}
        }
    }
]

for doc in courses_collection.aggregate(pipeline2):
    print(doc)
```

```python
# Pecah Studen koleksi agar bisa di Join
students_collection = db['students']
students_collection.delete_many({})  # bersihkan dulu

students_data = [
    {'student_id': 1, 'name': 'Asep'},
    {'student_id': 2, 'name': 'Boby'},
    {'student_id': 3, 'name': 'Chunha'},
    {'student_id': 4, 'name': 'David'},
    {'student_id': 5, 'name': 'Eva'},
    {'student_id': 6, 'name': 'Fiona'},
    {'student_id': 10, 'name': 'Rusdi'}
]

students_collection.insert_many(students_data)
print("\nKoleksi 'students' berhasil dibuat.")
```

```python
print("\n=== TUGAS 3: Gabung courses dengan students menggunakan $lookup ===")
pipeline3 = [
    {'$unwind': '$students'},  # pecah list students
    {
        '$lookup': {
            'from': 'students',           # koleksi tujuan
            'localField': 'students.student_id',  # field di courses
            'foreignField': 'student_id',         # field di students
            'as': 'student_info'                  # hasil join
        }
    },
    {'$unwind': '$student_info'},  # ambil detail siswa
    {
        '$project': {
            '_id': 0,
            'course': 1,
            'department': 1,
            'student_name': '$student_info.name'
        }
    }
]

for doc in courses_collection.aggregate(pipeline3):
    print(doc)
```
