```python
from pymongo import MongoClient

client = MongoClient("mongodb://localhost:27017/")

db = client["company_db"]
collection = db["employees"]
print("✅ Connected to MongoDB")
```

```python
# Tugas 2
# Contoh Insert Data
employee_data = {
    'name' : 'Rusdi',
    'department' : 'Finance',
    'age' : 30,
    'Salary' : 5000
}
collection.insert_one(employee_data)
print("Data Inserted")

# Insert beberapa data sekaligus (multiple insert)
multiple_employees = [
    {'name': 'Agus', 'department': 'IT', 'age': 32, 'salary': 5500},
    {'name': 'Rita', 'department': 'HR', 'age': 27, 'salary': 4200},
    {'name': 'Diana', 'department': 'Finance', 'age': 35, 'salary': 6000},
    {'name': 'Gatot', 'department': 'IT', 'age': 25, 'salary': 4800}
]
collection.insert_many(multiple_employees)
print("Beberapa data berhasil dimasukkan!")
print("=== Data Pegawai ===")
for emp in collection.find():
    print(emp)
```

```python
# Update banyak dokumen
collection.update_many(
    {'department': 'IT'},
    {'$mul': {'salary': 1.10}}  # Naik 10%
)
print("Gaji pegawai di IT naik 10%!")

# Tampilkan hasil setelah update
for emp in collection.find({'department': 'IT'}):
    print(emp)
```

```python
collection.update_one(
    {'name': 'Rusdi'},
    {'$set': {'name': 'Rusdiana'}}
)
print("Nama Rusdi diupdate menjadi Rusdiana.")
```

```python
print("=== Data Pegawai ===")
for emp in collection.find():
    print(emp)
```

```python
# Hapus satu dokumen
collection.delete_one({'name': 'Rita'})
print("Data Rita dihapus")
```

```python
# Hapus banyak dokumen
collection.delete_many({'department': 'Finance'})
print("Semua pegawai Finance dihapus.")
```

```python
print("=== Data Setelah Delete ===")
for emp in collection.find():
    print(emp)
```

```python
# Tugas 3
pipeline = [
    {'$group' : {'_id': '$department', 'average_salary': {'$avg': '$salary'}}}
]
for result in collection.aggregate(pipeline):
    print(result)
```

```python
from pymongo import MongoClient
import pandas as pd

# Koneksi MongoDB lokal
client = MongoClient("mongodb://localhost:27017/")

# Pilih database
db = client["toko_online"]

# Buat koleksi baru untuk produk
products = db["products"]

print("Koneksi ke database 'toko_online' berhasil!")
```

```python
# Hapus data lama agar tidak dobel (opsional)
products.delete_many({})

# Masukkan data produk
produk_list = [
    {"nama": "Laptop Asus", "kategori": "Elektronik", "harga": 8500000, "stok": 10},
    {"nama": "Smartphone Samsung", "kategori": "Elektronik", "harga": 6500000, "stok": 15},
    {"nama": "Headphone JBL", "kategori": "Elektronik", "harga": 1200000, "stok": 20},
    {"nama": "Kemeja Pria", "kategori": "Fashion", "harga": 250000, "stok": 30},
    {"nama": "Celana Jeans", "kategori": "Fashion", "harga": 350000, "stok": 25},
    {"nama": "Sepatu Nike", "kategori": "Fashion", "harga": 950000, "stok": 18},
    {"nama": "Buku Pemrograman Python", "kategori": "Buku", "harga": 150000, "stok": 40},
    {"nama": "Buku Filsafat", "kategori": "Buku", "harga": 175000, "stok": 35},
    {"nama": "Snack Ring", "kategori": "Makanan", "harga": 8000, "stok": 100},
    {"nama": "Coklat Silverqueen", "kategori": "Makanan", "harga": 25000, "stok": 80}
]

# Insert data ke koleksi
products.insert_many(produk_list)
print("✅ 10 produk berhasil dimasukkan!")
```

```python
# Hitung rata-rata harga produk
avg_price = list(products.aggregate([
    {"$group": {"_id": None, "rata_rata_harga": {"$avg": "$harga"}}}
]))[0]["rata_rata_harga"]

print(f"Rata-rata harga produk: Rp{avg_price:,.0f}")
```

```python
print("=== Produk dengan harga di atas rata-rata ===")
for p in products.find({"harga": {"$gt": avg_price}}):
    print(p)
```

```python
# Tampilkan di DataFrame Pandas
data_df = pd.DataFrame(list(products.find({"harga": {"$gt": avg_price}}, {"_id": 0})))
data_df
```

```python
pipeline = [
    {"$group": {"_id": "$kategori", "total_produk": {"$sum": 1}}},
    {"$sort": {"total_produk": -1}}
]

hasil = list(products.aggregate(pipeline))

print("=== Total Produk per Kategori ===")
for h in hasil:
    print(f"{h['_id']}: {h['total_produk']} produk")

```

```python
print("=== Semua Data Pegawai ===")
for emp in collection.find({}, {"_id": 0}):
    print(emp)

```

```python
# Tugas Tambahan
# Tugas 1
pipeline_top5 = [
    {"$sort": {"department": 1, "salary": -1}},  # Urutkan berdasarkan departemen & gaji tertinggi
    {"$group": {
        "_id": "$department",
        "top5": {"$push": {"nama": "$name", "gaji": "$salary", "umur": "$age"}},
    }},
    {"$project": {
        "_id": 1,
        "top5": {"$slice": ["$top5", 5]}  # Ambil 5 teratas per departemen
    }}
]

top5_result = list(collection.aggregate(pipeline_top5))

print("=== 5 Karyawan dengan Gaji Tertinggi per Departemen ===")
for dept in top5_result:
    print(f"\nDepartemen: {dept['_id']}")
    for emp in dept["top5"]:
        print(f"  - {emp['nama']} | Gaji: {emp['gaji']} | Umur: {emp['umur']}")
```

```python
data_top5 = []
for dept in top5_result:
    for emp in dept["top5"]:
        data_top5.append({
            "Departemen": dept["_id"],
            "Nama": emp["nama"],
            "Gaji": emp["gaji"],
            "Umur": emp["umur"]
        })

pd.DataFrame(data_top5)
```

```python
# Tugas 2
# Hitung dulu berapa yang akan dihapus
count_under30 = collection.count_documents({"age": {"$lt": 30}})
print(f"Ada {count_under30} karyawan berusia di bawah 25 tahun.")

# Hapus karyawan dengan usia < 30
collection.delete_many({"age": {"$lt": 30}})
print("Semua karyawan di bawah 30 tahun berhasil dihapus!")

# Verifikasi
for emp in collection.find({}, {"_id": 0, "name": 1, "age": 1}):
    print(emp)
```

```python
# Tugas 3
pipeline_report = [
    {"$group": {
        "_id": "$department",
        "total_gaji": {"$sum": "$salary"},
        "rata_umur": {"$avg": "$age"},
        "jumlah_karyawan": {"$sum": 1}
    }},
    {"$sort": {"_id": 1}}
]

laporan = list(collection.aggregate(pipeline_report))

print("=== Laporan Ringkas Gaji & Umur per Departemen ===")
for row in laporan:
    print(f"Departemen: {row['_id']}")
    print(f"  Total Gaji     : {row['total_gaji']}")
    print(f"  Rata-rata Umur : {row['rata_umur']:.2f}")
    print(f"  Jumlah Karyawan: {row['jumlah_karyawan']}\n")
```
